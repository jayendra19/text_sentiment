```python
# Define ranges for categorization
negative_range = (1, 2.99)
neutral_range = (3, 3.99)
positive_range = (4, 5)
```

```python
# Custom function to categorize labels
def categorize_label(label):
    if negative_range[0] <= label <= negative_range[1]:
        return 'negative'
    elif neutral_range[0] <= label <= neutral_range[1]:
        return 'neutral'
    elif positive_range[0] <= label <= positive_range[1]:
        return 'positive'
```
